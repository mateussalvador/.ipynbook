# Decoradores

- Decoradores so funes que embrulham outras funes. Ex.:
```python
@decorador
def funo():
        ...
```
- Criam um comportamento diferente para a funo que est sendo deracorada embaixo do syntax sugar (`@decorador`)

## Funes como objetos
- Em Python, funes so objetos. Assim como int e list tambm so objetos.
- Isso significa que podemos atribuir funes a variveis e coloc-las em contineres.
### Exemplo
- Sendo assim, podemos atribuir a variveis, armazenar em listas, entre outras coisas.

````python
def soma(x, y):
    return x + y

sominha = soma
print("Sominha:", sominha(1, 2))

somo = soma
print("Somo:", somo(543.20, 87539.433553632))

l_funcs = [soma, ]
print("Chamando a funo de dentro de uma lista de funes:" ,l_funcs[0](1, 5))
````

### Exemplo 2

````python
def soma(x, y):
    return x + y

def sub(x, y):
    return x - y

def mul(x, y):
    return x * y

def div(x, y):
    return x / y

def calculadora(op, x, y):
    operaes = {
        "+": soma,
        "-": sub,
        "*": mul,
        "/": div
    }
    return operaes[op](x, y)

print(calculadora("*", 5, 25))
````

# Funes de ordem superior
- So funes que podem receber ou retornar outra funo

### Exemplo recebendo funo como parmetro

````python
from functools import reduce

def soma(x, y):
    return x + y

resultado = reduce(soma, [1, 2, 3, 4, 5])

print(resultado)
````

### reduce
-  uma funo de functools, que reduz uma lista de valores a um s valor, usando como base uma funo especificada. No exemplo acima, ele usou a funo soma e foi somando os pares at chegar no resultado **15**.
- Note que ela recebe uma funo como um dos argumentos.

### Exemplo retornando funo

````python
from functools import partial

def soma(x, y):
    return x + y

soma_1 = partial(soma, 1)
soma_10 = partial(soma, 10)

print(soma_1(10))
print(soma_10(10))
````

### partial
- Uma funo que retorna uma determinada funo com um parmetro deles j 'pr-definido'

### Terceiro exemplo de HOF

````python
def soma(x, y):
    return x + y

def zip_with(func, iter_a, iter_b):
    return list(map(func, iter_a, iter_b))

print(zip_with(soma, [1, 2, 3], [1, 2, 3]))
````

### map
- O map  uma funo que recebe uma funo como argumento, e recebe um ou mais iterveis. O seu objetivo  percorrer o itervel aplicando aquela funo a cada um dos elementos e gerando uma nova lista.
- Caso a funo tenha dois parmetros, deveremos ter dois iterveis. Se a funo tiver trs, precisamos de trs iterveis, assim sucessivamente.

````python
lista = [2, 5, 8]
lista2 = [3, 6, 9]

print(list(map(lambda x, y: (x - y) * 2, lista, lista2)))
# 2 - 3 = -1 -> -1 * 2 = -2
# 5 - 6 = -1 -> -1 * 2 = -2
# 8 - 9 = -1 -> -1 * 2 = -2
````

### Um exemplo mais insano

````python
from functools import partial

def zip_with_2(func):
    return partial(map, func)

zip_soma = zip_with_2(soma)
zip_mul = zip_with_2(mul)

# Testando zip_soma
print(
    list(
        zip_soma(
            [1, 2, 3],
            [1, 2, 3]
        )
    )
)

# Testando zip_mul
print(
    list(
        zip_mul(
            [1, 2, 3],
            [1, 2, 3]
        )
    )
)
````

# Closures e funes aninhadas
- Funes aninhadas so funes definidas dentro de funes

````python
def ola(nome):
    def func_interna(nome):
        if nome.lower() == "marilene":
            print(f"Ol {nome}. A noite, tainha, vinho e muito ...")
        else:
            print(f"Ol {nome}, boas vindas!")
    func_interna(nome)

ola("Marilene")
````

## Pra que servem funes aninhadas?
1. Funes ajudantes
    - Evitar repetir cdigo dentro da funo
    - Colocar toda a complexidade em um lugar s
2. Encapsulamento
    - A funo interna no pode ser acessada globalmente
3. Escopo de varivels
    - A funo interna pode usar variveis da funo externa

# Closures
- So funes que encapsulam uma funo interna e a retornam.
- O grande passo das closures  que seu escopo  preservado. Ou seja, as variveis de "fora" da funo interna tambm so preservados.

## Variveis livres
- So as variveis de escopo `nonlocal`

# Decoradores
- Uma closure pode ser um decorador **se e somente se:**
    - A funo externa receber **uma (e somente uma)** funo
        - A funo recebida  a varivel livre
        - Retorna a funo interna

### Exemplo

````python
def decorador(func):
    def interna(*args):
        resultado = func(*args)
        return f"Sou uma closure e sua funo retornou {resultado}"
    return interna

def soma(x, y):
    return x + y

decorada = decorador(soma)
decorada(1, 2)
````

--- 
- Como dito anteriormente, **os decoradores so acar sinttico para closures**

````python
# Sem acar sinttico
def decorador(func):
    def interna(*args):
        resultado = func(*args)
        return f"Sou uma closure e sua funo retornou {resultado}"
    return interna

def soma(x, y):
    return x + y

# No prximo exemplo, vamos remover a linha `decorada = decorador(soma)`
decorada = decorador(soma)
# decorada(1, 2)
````

````python
# Com acar sinttico
def decorador(func):
    def interna(*args):
        resultado = func(*args)
        return f"Sou uma closure e sua funo retornou {resultado}"
    return interna

# Inserindo o acar sinttico
@decorador
def soma(x, y):
    return x + y

soma(1, 2)
````

- Basicamente, ele evita que tenhamos que passar a funo decorada para dentro de uma varivel para que possamos utilizar.
- **Agora, basta chamar a prpria funo que foi decorada e, automaticamente, o Python ir fazer o servio de pass-la como argumento para a funo decoradora (que foi marcada com o acar sinttico).**
- Mais diretamente ainda, a funo soma ao ser decorada com `@`, deixa de existir. Ela passa a ser `interna`

### Outro exemplo

````python
def decorador(func):
    def interna(*args):
        resultado = func(*args)
        return f"Sou uma closure e sua funo retornou {resultado}"
    return interna

@decorador
def inverte_texto(text):
    return text[::-1].lower().capitalize()

inverte_texto("Mateus")
````

## Um exemplo 'real' de decoradores

````python
from datetime import datetime

def medidor_de_tempo(func):
    
    def aninhada(*args, **kwargs):
        tempo_inicial = datetime.now()
        
        resultado = func(*args, **kwargs)
        
        tempo_final = datetime.now()
        
        tempo = tempo_final - tempo_inicial
        
        print(f"{func.__name__} demorou {tempo.total_seconds()} segundos.")

        return resultado

    return aninhada

@medidor_de_tempo
def somar_ate_100():
    res = 0
    for num in range(101):
        res += num
    return res

print(somar_ate_100())
````

# Decoradores
- Um decorador  um 'chamvel' que recebe um alvo e devolve o objeto que ficar no lugar dele. Para funes, o padro  "recebe funo, retorna funo".

## Acar sinttico "`@`"
- Equivale a uma atribuio feita no momento da definio
- `@dec` em cima de `f` equivale a `f = dec(f)` logo aps a funo ser criada.
- Ou seja, ao escrever
```python
@meu_decorador
def minha_funcao():
    print("Executando a funo original.")
```
- Na verdade, estamos escrevendo:
```python
def minha_funcao():
    print("Executando a funo original.")

minha_funcao = meu_decorador(minha_funcao)
```

## Decoradores com argumentos
- Vamos pensar num decorador `@repetir(3)` que executa uma funo 3 vezes.
- Para isso, precisamos de mais um nvel de aninhamento. Criamos uma funo externa que aceita os argumentos e retorna o decorador.
- Essa abordagem  chamada de **Fbrica de Decoradores**.

### Estrutura
1. **Fbrica (`repetir`)**: (aceita os argumentos `num_vezes`)
2. **Decorador Real (`decorador`)** O que a fbrica retorna. ELe aceita a funo a ser decorada (`funcao_original`)

3. **Wrapper (`wrapper`)**: A funo que substitui a original. TEm acesso tanto aos argumentos da fbrica (`num_vezes`) quanto aos da funo original (`*args, **kwargs`)

````python
import functools

def repetir(num_vezes):
    """Fbrica que cria e retorna o decorador."""
    def decorador(funcao_original):
        """O decorador real que recebe a funo."""
        @functools.wraps(funcao_original) # (Mais sobre isso depois)
        def wrapper(*args, **kwargs):
            """O wrapper que executa a lgica."""
            print(f"--- Repetindo {num_vezes} vezes ---")
            for i in range(num_vezes):
                print(f"Execuo {i + 1}:")
                resultado = funcao_original(*args, **kwargs)
            return resultado
        return wrapper
    return decorador

@repetir(3)
def saudar(nome):
    """Imprime uma saudao simples."""
    print(f"OL, {nome}!")

saudar("Maria")
````

- A chamada `@repetir(3)`  traduzida para `funcao = repetir(3)(funcao)`
